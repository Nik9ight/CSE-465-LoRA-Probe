## 5. Visualization
Display some of the generated results.

```python
import matplotlib.pyplot as plt
import glob
import matplotlib.image as mpimg

def show_results(model_name, mode):
    images = glob.glob(f"./results/{model_name}_{mode}_query_*.png")
    for img_path in images[:3]: # Show first 3
        img = mpimg.imread(img_path)
        plt.figure(figsize=(20, 5))
        plt.imshow(img)
        plt.axis('off')
        plt.show()

print("ResNet-101 Baseline Results")
show_results('resnet101', 'baseline')

print("ResNet-101 LoRA Results")
show_results('resnet101', 'lora')
```

```python
print("GoogLeNet Baseline Results")
show_results('googlenet', 'baseline')

print("GoogLeNet LoRA Results")
show_results('googlenet', 'lora')
```
